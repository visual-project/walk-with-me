```python
import numpy as np
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sns

fm = mp.font_manager.FontManager()
# 한글지원하는 폰트명으로 재할당
plt.rcParams['font.family'] = 'Malgun Gothic'
#plt.rc('font',family='Malgun Gothic')
#plt.rc('axes',unicode_minus=False)
```

```python
park = pd.read_csv(r'C:\Users\User\Desktop\datastudy\project\first_project\walk-with-me\raw/공원 데이터 정리.csv', sep =',',encoding='UTF8')
park = park.iloc[:,[0,1,3,4,5]]
park['공원면적'] = park['공원면적'].str.replace(',', '').astype('float')
park['공원 수'] = park['공원 수'].str.replace(',', '').astype('float')
park.dtypes

tree = pd.read_csv(r'C:\Users\User\Desktop\datastudy\project\first_project\walk-with-me\raw/도시림 데이터 정리.csv', sep =',',encoding='UTF8')
#park['공원면적'] = park['공원면적'].str.replace(',', '').astype('float')
#park['공원 수'] = park['공원 수'].str.replace(',', '').astype('float')

tree.dtypes
```

```python
pt_join = pd.DataFrame(pd.merge(left = park , right = tree, how = "inner", on = "자치구"))
pt_join.drop(index=[0], axis=0, inplace=True) # 행삭제
pt_join
```

```python
health = pd.read_csv(r'C:\Users\User\Desktop\datastudy\project\first_project\walk-with-me\raw\지역사회건강조사 (전처리된 데이터)_2019/통합 테이블_2019.csv', sep =',',encoding='UTF8')
health.rename(columns = {'지역':'자치구'}, inplace = True) # 특정 컬럼명 변경
health
```
