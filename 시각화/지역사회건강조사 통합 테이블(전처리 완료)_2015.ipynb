```python
import pandas as pd
import os
 
path_dir = '../raw/지역사회건강조사 (전처리된 데이터)_2015/'
 
file_list = os.listdir(path_dir)
data_list = []
for i in file_list:
    if i.split(' ')[0] == '지역사회건강조사':
        data_list.append(pd.read_csv('../raw/지역사회건강조사 (전처리된 데이터)_2015/%s' % i))

for i in data_list:
    i.set_index(['지역','연령별','성별'],inplace=True)
```

```python
df_health = pd.concat(data_list, axis=1, join='inner')
```
