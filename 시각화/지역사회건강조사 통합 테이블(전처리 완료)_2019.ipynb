```python
import pandas as pd

data1 = pd.read_csv('../raw/지역사회건강조사 (전처리된 데이터)_2019/지역사회건강조사 (건강수준)_2019.csv')
data2 = pd.read_csv('../raw/지역사회건강조사 (전처리된 데이터)_2019/지역사회건강조사 (걷기)_2019.csv')
data3 = pd.read_csv('../raw/지역사회건강조사 (전처리된 데이터)_2019/지역사회건강조사 (격렬한 신체활동)_2019.csv')
data4 = pd.read_csv('../raw/지역사회건강조사 (전처리된 데이터)_2019/지역사회건강조사 (중등도 신체활동)_2019.csv')
#data5 = pd.read_csv('../raw/지역사회건강조사 (전처리된 데이터)_2019/')
data6 = pd.read_csv('../raw/지역사회건강조사 (전처리된 데이터)_2019/지역사회건강조사 (우울감)_2019.csv')
data7 = pd.read_csv('../raw/지역사회건강조사 (전처리된 데이터)_2019/지역사회건강조사 (스트레스)_2019.csv')
```

```python
list_df = [data1,data2,data3,data4,data6,data7]
for i in list_df:
    i.set_index(['지역','연령별','성별'],inplace=True)
```

```python
df = pd.concat([data1,data2,data3,data4,data6,data7],axis=1, join='inner')
```
