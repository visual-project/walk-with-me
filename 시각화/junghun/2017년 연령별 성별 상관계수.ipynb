```python
import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
fm = mp.font_manager.FontManager()
plt.rcParams['font.family'] = 'Malgun Gothic'

import pandas as pd
list_age = ['계','19~44','45~64','65~74','75 이상']
list_sex = ['계','남','여']
array = {}
for z in range(0,9):
    for i in list_age:
        for j in list_sex:
            data1 = pd.read_csv('../../raw/통합 테이블(공원)_2017.csv')
            data2 = pd.read_csv('../../raw/지역사회건강조사 (전처리된 데이터)_2017/통합 테이블(건강)_2017.csv')
            data2.rename(columns={'지역':'자치구'},inplace = True)
            df2 = data2[data2['연령별'] == i]
            df2 = df2[df2['성별'] == j]
            df2.set_index('자치구',inplace = True)
            data1.set_index('자치구',inplace = True)
            df = pd.concat([df2,data1],axis=1)
            df = df.drop(df.index[0])
            df = df.drop(['연령별','성별'],axis=1)
            df1 = df[[df.columns[z],'1인당 생활권 도시림 면적']]
            #df1.rename(columns={df.columns[z]:'_'+ df.columns[z].replace(' ','_'),'1인당 생활권 도시림 면적':'_1인당생활권도시림면적'},inplace = True)
            corr = df1.corr()
            array[df.columns[z]+i+j] = corr.iloc[0][1]
```
