```python
import pandas as pd
import numpy as np

data_raw = pd.read_csv('../raw/지역사회건강조사 원시자료/chs19_a.txt',sep ='\t')
```

```python
data_pbhlth_code = pd.read_excel('../raw/지역사회건강조사 2019년 원시자료 이용지침서 등/지역사회건강조사 2019 원시자료 참고사항_보건소정보, 세대유형.xlsx',header=3)
data_pbhlth_code = data_pbhlth_code[data_pbhlth_code['시도코드'] == 11][['보건소코드','보건소명']]

data_pbhlth_code["보건소명"] = data_pbhlth_code["보건소명"].str.replace('보건소', '', regex=True)
data_pbhlth_code.columns=['보건소코드','자치구']
data_pbhlth_code.set_index('보건소코드',inplace=True)
data_pbhlth_code.to_dict()
data_pbhlth_code.to_dict()['자치구']
```

```python
df_raw1 = data_raw[['age','sex','pbhlth_code','qoa_01z1','pha_04z1','pha_05z1','pha_06z1','pha_07z1','pha_08z1','pha_09z1','phb_01z1','phb_02z1','phb_03z1','obb_01z1','mta_01z1','mtb_01z1','qoc_07z1','ena_01d1','ena_01e1']]
index = ['만나이','성별','보건소번호','주관적 건강수준','격렬한 신체활동 일수','격렬한 신체활동 시간(시)','격렬한 신체활동 시간(분)','중등도 신체활동 일수','중등도 신체활동 시간(시)','중등도 신체활동 시간(분)','걷기 실천 일수','걷기 실천 시간(시)','걷기 실천 시간(분)','체중조절 경험 여부','주관적 스트레스 수준','우울감 경험 여부','행복감 지수','사회환경_동네 자연환경','사회환경_동네 생활환경']
df_raw1.columns = index
df_raw2 = df_raw1.copy() # SettingWithCopyWarning 없애려고 넣음..없어도 돌아는 감
df_raw2['자치구'] = df_raw2['보건소번호'].map(data_pbhlth_code.to_dict()['자치구'])
df_raw2 = df_raw2.drop(['보건소번호'],axis=1)
df_raw2['성별'].replace({1:'남',2:'여'}, inplace = True)
df_raw2
```

```python
# 생애주기 구분
def func(x) :
    if x <= 44 :
        return "19-44세"
    elif x <= 64 : 
        return "45-64세"
    elif x <= 74 : 
        return "65-74세"
    else:
        return "75세 이상"

df_raw2["생애주기"] = df_raw2["만나이"].apply(lambda x : func(x))
```

```python
df_activity = df_raw2.copy()[['자치구','만나이', '생애주기', '성별', '격렬한 신체활동 일수', '격렬한 신체활동 시간(시)',
       '격렬한 신체활동 시간(분)', '중등도 신체활동 일수', '중등도 신체활동 시간(시)', '중등도 신체활동 시간(분)',
       '걷기 실천 일수', '걷기 실천 시간(시)', '걷기 실천 시간(분)']]
```

```python
## 격렬한 운동 관련
# 시간 --> 분으로 환산
def transtime(df_activity) :
    if 0 <= df_activity['격렬한 신체활동 시간(시)'] <= 24 and df_activity['격렬한 신체활동 시간(분)'] <= 59 :
        return df_activity['격렬한 신체활동 시간(시)']*60 + df_activity['격렬한 신체활동 시간(분)']
    elif df_activity['격렬한 신체활동 시간(시)'] > 24 and df_activity['격렬한 신체활동 시간(분)'] <= 59 :
        return df_activity['격렬한 신체활동 시간(분)']
    else :
        return 0

df_activity["격렬_시분_환산"] = df_activity.apply(transtime, axis =1)
print("격렬한 운동 시분_환산 합: ",df_activity["격렬_시분_환산"].sum())

# 지표 환산
def health_index(df_activity) :
    if df_activity['격렬한 신체활동 일수'] < 3 :
        return 0
    elif 3 <= df_activity['격렬한 신체활동 일수'] and df_activity['격렬_시분_환산'] >= 20 :
        return 1
    else :
        return 0

df_activity['격렬한 신체활동 실천'] = df_activity.apply(health_index, axis =1)
print('격렬한 신체활동 실천 해당자 합 : ', df_activity['격렬한 신체활동 실천'].sum())
    
# 응답자수(결측치 제외)
def valid_response(df_activity) :
    if df_activity['격렬한 신체활동 일수'] in  [77, 99] :
        return 0
    elif df_activity['격렬한 신체활동 시간(시)'] in [77, 99] :
        return 0
    elif df_activity['격렬한 신체활동 시간(분)'] in [77, 99] :
        return 0
    else :
        return 1
    
df_activity['격렬한 신체활동 응답'] = df_activity.apply(valid_response, axis =1)
print('격렬한 신체활동 응답자 합 : ', df_activity['격렬한 신체활동 응답'].sum())
```

```python
## 중등도 운동 관련
# 시간 --> 분으로 환산
def transtime(df_activity) :
    if 0 <= df_activity['중등도 신체활동 시간(시)'] <= 24 and df_activity['중등도 신체활동 시간(분)'] <= 59 :
        return df_activity['중등도 신체활동 시간(시)']*60 + df_activity['중등도 신체활동 시간(분)']
    elif df_activity['중등도 신체활동 시간(시)'] > 24 and df_activity['중등도 신체활동 시간(분)'] <= 59 :
        return df_activity['중등도 신체활동 시간(분)']
    else :
        return 0

df_activity["중등_시분_환산"] = df_activity.apply(transtime, axis =1)
print("중등도 운동 시분_환산 합: ",df_activity["중등_시분_환산"].sum())

# 지표 환산
def health_index(df_activity) :
    if df_activity['중등도 신체활동 일수'] < 5 :
        return 0
    elif 5 <= df_activity['중등도 신체활동 일수'] and df_activity['중등_시분_환산'] >= 30 :
        return 1
    else :
        return 0

df_activity['중등도 신체활동 실천'] = df_activity.apply(health_index, axis =1)
print('중등도 신체활동 실천 해당자 합 : ', df_activity['중등도 신체활동 실천'].sum())

# 응답자수(결측치 제외)
def valid_response(df_activity) :
    if df_activity['중등도 신체활동 일수'] in  [77, 99] :
        return 0
    elif df_activity['중등도 신체활동 시간(시)'] in [77, 99] :
        return 0
    elif df_activity['중등도 신체활동 시간(분)'] in [77, 99] :
        return 0
    else :
        return 1
    
df_activity['중등도 신체활동 응답'] = df_activity.apply(valid_response, axis =1)
print('중등도 신체활동 응답자 합 : ', df_activity['중등도 신체활동 응답'].sum())
```

```python
df_pv_activity =df_activity.pivot_table(index=['자치구','생애주기','성별'],values = ['격렬한 신체활동 실천','격렬한 신체활동 응답','중등도 신체활동 실천','중등도 신체활동 응답'],
                        aggfunc=np.sum,margins=True,margins_name='서울시')

df_pv_activity['격렬한 신체활동 실천율'] = round(df_pv_activity['격렬한 신체활동 실천']/df_pv_activity['격렬한 신체활동 응답']*100,1)
df_pv_activity['중등도 신체활동 실천율'] = round(df_pv_activity['중등도 신체활동 실천']/df_pv_activity['중등도 신체활동 응답']*100,1)
#df_pv_activity.columns
df_pv_activity = df_pv_activity[['격렬한 신체활동 실천', '격렬한 신체활동 응답', '격렬한 신체활동 실천율','중등도 신체활동 실천', '중등도 신체활동 응답', '중등도 신체활동 실천율']]
df_pv_activity.head(50)
```
