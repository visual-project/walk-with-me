```python
import pandas as pd

df_population = pd.read_csv('../raw/서울시 주민등록인구 (연령별구별) 통계.txt', thousands = ',', sep ='\t',encoding='UTF8',)
```

```python
df_population.info()
```

```python
df_population.tail()
```

```python
df_population.drop(['기간'], axis='columns', inplace=True) 
```

```python
df_population.replace('-', '0', inplace = True)
df_population.replace('', '0', inplace = True)
```

```python
df_population.tail()
```

```python
df_population['85~89세'] = df_population['85~89세'].str.replace(',', '').astype('int64')
df_population['90~94세'] = df_population['90~94세'].str.replace(',', '').astype('int64')
df_population['95~99세'] = df_population['95~99세'].str.replace(',', '').astype('int64')
df_population['100세 이상'] = df_population['100세 이상'].str.replace(',', '').astype('int64')
```

```python
df_population.dtypes
```

```python
df_population_total = df_population.copy()[df_population['구분']=='계']
```

```python
df_population_total.head()
```

```python
df_population_total.drop(['구분'], axis='columns', inplace=True) 
df_population_total.head()
```

```python
print(df_population_total['행정구역별'].unique())
```

```python
df_population_total['행정구역별'].replace('합계','서울시', inplace = True)
print(df_population_total['행정구역별'].unique())
```

```python
df_population_total.head()
```

```python
df_population_total['10대 이하'] = df_population_total['0~4세'] + df_population_total['5~9세'] + df_population_total['10~14세'] + df_population_total['15~19세']
df_population_total['19-44'] = df_population_total['20~24세'] + df_population_total['25~29세'] + df_population_total['30~34세'] + df_population_total['35~39세'] + df_population_total['40~44세']
df_population_total['45-64'] = df_population_total['45~49세'] + df_population_total['50~54세'] + df_population_total['55~59세'] + df_population_total['60~64세']
df_population_total['65-74'] = df_population_total['65~69세'] + df_population_total['70~74세']
df_population_total['70대 이상'] = df_population_total['75~79세'] + df_population_total['80~84세'] + df_population_total['85~89세']+df_population_total['90~94세'] + df_population_total['95~99세']+df_population_total['100세 이상']
```

```python
cols = df_population_total.columns
df_population_total.drop(cols[2:23], axis = 1,inplace = True)
```

```python
df_population_total.head()
```

```python
df_population_total.rename(columns={'행정구역별':'자치구', '계':'주민등록인구'}, inplace=True)
df_population_total.set_index('자치구',inplace=True)
```

```python
df_population_total
```

```python
df_population_total.to_csv("../raw/서울시 주민등록인구 (연령별구별) 통계_2019_Q4.csv", mode='w',encoding = 'utf-8-sig')
```
