```python
import pandas as pd
import numpy as np


data1 = pd.read_excel('../raw/전국도시공원정보표준데이터-20210706.xls',header=1)
```

```python
data1
```

```python
df1 = data1.drop(['소재지도로명주소','소재지지번주소','데이터기준일자','제공기관코드','공원보유시설(운동시설)','공원보유시설(유희시설)','공원보유시설(편익시설)',
                 '공원보유시설(교양시설)','공원보유시설(기타시설)','지정고시일','관리기관명','전화번호'],axis=1)
```

```python
df1['시도'] = df1.제공기관명.str.split(' ').str[0]
df1['자치구'] = df1.제공기관명.str.split(' ').str[1]
```

```python
df1
```

```python
is_seoul = df1['시도'] == '서울특별시'
df2 = df1[is_seoul]
```

```python
df2_locate=df2[['관리번호','공원명','위도','경도','공원면적','자치구']]
df2_locate
```

```python
df2_area=df2[['공원면적','자치구']].groupby('자치구')
df222= df2_area.sum()
area_sum=round(df222['공원면적'].sum(),2)
df222
```

```python
df222.loc['서울시']=area_sum
df222
```

```python
freq = df2['자치구'].value_counts() 
freq
```

```python
df3 = pd.DataFrame(freq)
```

```python
df3 = df3.reset_index()
df3.columns=['자치구','공원수1']
```

```python
df3.set_index('자치구',inplace=True)
```

```python
num_sum=df3['공원수1'].sum()
df3.loc['서울시']=num_sum
df3
```

```python
df3.to_csv("../raw/자치구별 공원수.csv", mode='w',encoding = 'utf-8-sig')
df2_locate.to_csv("../raw/공원위치데이터.csv", mode='w',encoding = 'utf-8-sig')
df222.to_csv("../raw/자치구별 면적.csv", mode='w',encoding = 'utf-8-sig')
```
